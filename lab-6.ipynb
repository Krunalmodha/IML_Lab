**6 a) To convert a NumPy array to a Pandas series.**

```python
import pandas as pd 
import numpy as np

numpyArray = np.array(['q','a','e','v','f','s','w']) 
pandaSeries = pd.Series(numpyArray)
print("Pandas Series:",)
print(pandaSeries)
```

**6 b) To convert the first column of a DataFrame as a Series.**

```python
import pandas as pd 
 
dataFrame = pd.DataFrame({'Col 1':[11,22,33,299],'Col 2':[44,55,66,992]}) 

pandaSeries = dataFrame.iloc[:,0]
print(pandaSeries)
```

**6 c) To create the mean and standard deviation of the data of a given Series.**

```python
import pandas as pd

pandaSeries = pd.Series(data = [9,9,8,8,7,7,6,6,5,5,4,4,3,3,2,2,1,1])

print("Mean --> ",pandaSeries.mean())
print("Standard Deviation --> ",pandaSeries.std())
```

**6 d) To sort a given Series.**

```python
import pandas as pd
pandaSeries = pd.Series([110,500,256,49,21])

print("Ascending Order -->")
print(pandaSeries.sort_values(ascending = True))

print("\nDescending Order -->")
print(pandaSeries.sort_values(ascending = False))
```
