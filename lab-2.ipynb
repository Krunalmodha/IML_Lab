**2 a) To convert a list of numeric values into a 1 dimensional NumPY array**.

```python
import numpy

list = [20,33,54,86,200,70,9]

arrayy = numpy.array(list)

print ("List --> ", list)

print ("Array --> ", arrayy)
```

**2 b) To create a 3*3 matrix with values ranging from 2 to 10**.

```python
import numpy as np

arrayy = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])

reshapeArray = arrayy.reshape(3, 3)

print("Reshaped Array --> \n",reshapeArray)
```

**2 c) To append value at the end of array.**

```python
import numpy as np

arrayy = [54, 65, 23]

print("Original array --> ",arrayy)

arrayy = np.append(arrayy, [560])
print("After appending value at the end of the array --> ",arrayy)
```

**2 d) To create another shape from an array without changing its data(32 to 23)**

```python
import numpy as np

arrayX = np.array([1, 2, 3, 4, 5, 6])

arrayY = np.reshape(arrayX,(3,2))

print("Reshaped 3x2 --> \n",arrayY)

arrayZ = np.reshape(arrayX,(2,3))
print("Reshape 2x3 --> \n",arrayZ)
```
