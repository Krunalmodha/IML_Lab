**3 a) To split an array of 14 elements into 3 arrays,each with 2,4 and 8 elements in the original order.**

```python
import numpy as np
x = np.arange(1, 15)
print("Original array:\n",x)

print("After splitting:")

print(np.split(x, [2, 6]))
```

**3 b)To stack arrays horizontally (column wise).**

```python
import numpy as np

arrayOne = np.array([ 1, 2, 3] )
print ("1st Input array : \n", arrayOne)

arrayTwo = np.array([ 4, 5, 6] )
print ("2nd Input array : \n", arrayTwo)

result = np.hstack((arrayOne, arrayTwo))
print ("Output horizontally stacked array:\n ", result)
```
